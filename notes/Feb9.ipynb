```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(3,2, figsize=(10, 15))
tips = sns.load_dataset('tips')

sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', ax=ax[0,0])
ax[0,0].set_title('Total Bill vs Tip by Gender')

sns.countplot(data=tips, x='sex', ax=ax[0,1])
ax[0,1].set_title('Count by Gender')

sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', ax=ax[1,0])
ax[1,0].set_title('Total Bill vs Tip by Day')

sns.countplot(data=tips, x='sex', hue='day', ax=ax[1,1])
ax[1,1].set_title('Count by Gender by Day')

sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', ax=ax[2,0])
ax[2,0].set_title('Total Bill vs Tip by Time of Day')

sns.countplot(data=tips, x='sex', hue='time', ax=ax[2,1])
ax[2,1].set_title('Count by Gender by Time of Day')
```
