# The Nature of Data
Asking the right questions determines the data that you need to collect
- Does the past represent future?
- What do I need/want to model?
- How will the model be used?
- What data do I already have/need?
- How hard is it to gather the data?

# Big Data vs Smart Data
The bigger the data the longer it takes to model or train the model.
