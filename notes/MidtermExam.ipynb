# Midterm Exam
One of the goals of this class is to help you earn a Data Anaylsis certificate. To earn this certificate, you must demonstrate that you understand the principles and can use programming to create and analyze data. Because of this, projects and exams in this class will be geared toward that certificate.  
  
To demonstrate your understanding of Data Analysis principles, you will have 2 projects and 2 exams. The first project/exam pair will be considered your midterm exam. The other project/exam pair will be considered your final exam.

## Midterm Project
The project will be your opprotunity to demonstrate your ability to load, wrangle, and analyze data.
- Loading data
- Missing data
- Encoding data
- Statistics
- Graphing
- Analyzing data

Project
- You are getting an assignment from your boss
- Do an analysis and report back your findings
- 5-7 page report
    - PDF + code
    - iPython Notebook
