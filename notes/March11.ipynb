Unsurpervised: Labels not attached to the data
- Elections

To prevent overfitting you train your model with only 60-80% of the data and use the remaining 20-40% to test the model

## Statistical Learning

Most data follows a pattern in nature (a function)  
$$y=f(x) + \epsilon(epsilon)$$
epsilon is a random error term (noise) or confounding variables (outside influences)


$$\hat y = \hat f(x)$$
x is the data  
f^(x) is the model  
y^ is the prediction or the output of the model

when we try to model error terms it leads to overfitting

$$E[(y-\hat y)^2] = E[f(x) + \epsilon - \hat f(x)] \newline
    = [f(x) - \hat f(x)]^2 + var(\epsilon) $$

Reducible Error + irreducible error

There is a cap to how accurate our model will be
- 100% accurate model? Not possible because of irrducible errors.
