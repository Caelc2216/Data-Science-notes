## Definition of Data Science and a Data Scientist
- __Data Science__ is a study of ways to efficiently and effectively collect, wrangle, analyze, model, and draw conclusions from data
- A __Data Scientist__ ...
    - oversees the data science process
    - makes the decisions of what processes are useful and most beneficial
    - makes the decisions and conclusions at the end
    - adds personal experience and logic to the process
    - delivers an end result that adds calue to their employer
    - is a storyteller

CRISP-DM (Cross-Industry Standard Process for Data Mining)
1. Business understanding
2. Data understanding
3. Data preparation
    - Data handling/cleaning, Data Wrangling, EDA (Exploratory Data Analysis: Statistics, Visualization, Analysis) 
    - ETL (Extract, Transform, and Load)
4. Modeling
5. Evaluation
6. Deployment
    - Model monitoring and assessment

### Missing data
- Drop row
    - Only a few rows are missing
    - very little data in that row
    - Row obviously is bad
    - When critical data isn't lost if it's dropped
- Drop column
    - Column is virtually empty
    - Not necessary
    - When critical data isn't lost if it's dropped
- Imputing - Fill in the NaN with a value
    - average, median, ...
        - Average by category
    - Logical value 
        - e.g. look at credits taken to gyess grade level
        - e.g 'None' for no rating
    - Forward fill / Back fill / Interpolate
        - If the data is already ordered so it's logical
        - Sequence, such as time-based observations
        - Create a linear regression and fill in with that regression line (careful - make sure regression accurately represents the data and that correlation is high)
