### Statistical Learning (Machine Learning Models)

- Natural Functions
    - $y = f(x) + \epsilon$
- Model the natural function
    - $\hat y = \hat f(x) \approx f(x)$
    - Reducible Error   $|f(x) - \hat f(x)|$
    - Irreducible Error $\epsilon$
        - If we try to model $\epsilon$, we overfit our model
- Cross Validation (splitting data into two sets for training and testing the model)
- Feature Scaling
    |ID| Years of Experience | Salary |
    |:-:|:------------------:| :-----:|
    |1|4|$79,000|
    |2|1|$69,000|
    |Average| 4.5| $70,000

    $d_1 = \sqrt {(4 - 4.5)^2 + (79000 - 70000)^2} = 9000$  
    $d_2 = \sqrt {(1 - 4.5)^2 + (69000 - 70000)^2} = 1000$  
    - Min-Max Scaling (Normalization)
    - Normal distribution around mean (Standardization)
