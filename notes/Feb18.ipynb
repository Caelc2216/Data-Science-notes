## Feature Scaling

Employee records
| Time with company | Salary |
| :---------------: | :----: |
| 4                 | 62,000 |
| 8                 | 78,000 |
| 2                 | 62,000 |
| 3                 | 81,000 |


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
salaries = pd.DataFrame({
    'time' : [4, 8, 2, 3],
    'salary' : [62000, 78000, 62000, 81000]
})

fig, ax = plt.subplots(2,1, figsize=(10, 5))
sns.boxplot(data=salaries, x='time', ax=ax[0])
sns.boxplot(data=salaries, x='salary', ax=ax[1])

plt.tight_layout()
plt.show()
```

The idea of feature scaling is to put all features (columns of DF) onto the same scale  
Methods for Feature Scaling:
- Normalization (Min-Max Scaling) 
    - puts all values on a scale between 0 and 1
    $$x' = \frac{x-x_{min}}{x_{max}-x_{min}}$$
- Standardization (Z-score)
    - distance from the mean (in standatd deviations)
    - Typically between -3 and +3

    $$x' = \frac{x-\bar x}{s}$$

- Logarithmic

```python
def normalize(df):
    return (df - df.min()) / (df.max() - df.min())

salaries_norm = normalize(salaries)
```

```python
def standardize(df):
    return (df - df.mean()) / df.std(ddof=1)

salaries_std = standardize(salaries)
```
