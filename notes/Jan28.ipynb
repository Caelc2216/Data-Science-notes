## Missing Data
How is missing data represented?
- Null or Empty
- 0
- -1 or -9999 or 9999 {a ridiculous number that can't be true}
- Characters ("---", "N/A")
- NaN  
If we have missing values, what do we do with them?
- Drop
    - We can drop columns if data is missing for the majority of observations
    - We can drop rows if the missing data interfers with the model and if the other data isn't crucial
    - When to drop?
        - When the present data that would be dropped isn't crucial to the model
- Fill in missing values with an estimate (imputing)
    - Use a Linear Regression to estimate
    - Compare to other variables
    - Fill with average values
    - Interpolate (Good when values follow a sequence)
        - Average of surrounding vlaues
        - Forward Fill: A missing value takes on the previous value
        - Back FIll: A missing value takes on the next value


```python
import numpy as np
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
display(titanic)
```

```python
titanic.isna().sum()
```

```python
filter = (titanic['embarked'].isna())
titanic[filter]
```

```python
titanic.shape
```

```python
titanic.drop(['embarked', 'deck'], axis=1, inplace=True)
```

```python
titanic.isna().sum()
```

```python
titanic_filtered = titanic.drop([61,829], axis=0)
titanic_filtered[titanic_filtered['embark_town'].isna()]
titanic_filtered.loc[61]
```

```python
titanic.dropna(subset='embark_town', inplace=True)
```

When do I drop data?
- If more than half of a variable is missing, drop the variable
- If less than 10% of a variable is missing, drop the observations

Exact numbers up to Data Scientist's discretion  
Don't drop if the other information is still valuable
