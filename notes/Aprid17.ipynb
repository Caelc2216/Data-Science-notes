# Review of Statistical Learning

How do we model?  
$y = f(x) + \epsilon $  
  
We try to simulate f(x)  
$\hat y = \hat f(x)$  
- Can't include $\epsilon$ in our prediction of $\hat y$  
    - $\epsilon$ is random error
    - $|f(x) - \hat f(x)|$ indicates reducible error
    - $\epsilon$ is included in irreducible error
    - Trying to simulate $\epsilon$ will result in **overfitting**
- All we can do is model f(x)  
  
Sometimes, errors (including $\epsilon$) are emplified through modeling  
Because of the errors in models, it is easy to incorrectly use and to misinterpret models.

**Ethics** is an ongoing process of articulating values and, in turn, questioning and justifying decision based on those values
- Usually determined in terms of rights, obligations, benefits to society, or specific values  
  
**Ethical dilemmas** are situations where a difficult choice has to be made between different courses of action, each of which entails transgressing a moral principle.
- Any decision could harm one party or another
- Killing in self defense
- Civil rights  
What are some common ethical concerns?
- **Transparency**/Lack of understanding/lack of autonomy
- **Privacy** - Training on copyrighted or materials or personal information
- **Accountability to people** - Is AI bound by ethics?
- **AI Pseudoscience**
- **Security** - Can I use AI to break into systems? To steal people's information?
- **Unfair bias**
- **AI-driven unemployment and deskilling**
