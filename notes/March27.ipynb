How do we measure distance?  
**Pythagorean Theorem (Euclidean Distance or L2-Norm)**  
$d_x = a_x - b_x$  
$d_y = a_y - b_y$  

$d^2 = d_x^2 + d_y^2  

$d = \sqrt {d_x^2 + d_y^2}$  

In 3D  
$d = \sqrt {d_x^2 + d_y^2 + d_z^2}$  

in *n* dimension  
$L_2 = \sqrt{\sum_{i=1}^{n} d_i^2}$  

Is this the only way to measure distance?  
- In Manhattan, they measure blocks  

**L1-norm (Manhattan Distance)**  
$L_1 = \sum_i^n {|d_i|}$  

**L3-norm**  
$L_3 = \sqrt[3] {\sum_{i=1}^{n} {|d_i|}^3}$  
Where $d_i = a_i - b_i$  

**Lp-norm**  
$L_p = \sqrt[p] {\sum_{i=1}^{n} {|d_i|}^p}$  
We can visualize this for p = 1 and p = 1
- p = 1 gives the L1 form, or the Manhattan Distance
- p = 2 gives the L2 norm, or the Euclidean Distance (Pythagorean Therem)
We can't really visualize beyond that. But we can see the results numerically
  
**The Infinity Norm**  
$L_{\infty} = \lim_{p \to \infty} \sqrt[p]{(\sum_{i=1}^{n} |d_i|^p} = \max_i |d_i|$
The higher $p$ gets, the closer the norm gets to the largest component.
