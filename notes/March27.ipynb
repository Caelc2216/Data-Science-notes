How do we measure distance?  
**Pythagorean Theorem (Euclidean Distance or L2-Norm)**  
$d_x = a_x - b_x$  
$d_y = a_y - b_y$  

$d^2 = d_x^2 + d_y^2  

$d = \sqrt {d_x^2 + d_y^2}$  

In 3D  
$d = \sqrt {d_x^2 + d_y^2 + d_z^2}$  

in *n* dimension  
$L_2 = \sqrt{\sum_{i=1}^{n} d_i^2}$  

Is this the only way to measure distance?  
- In Manhattan, they measure blocks  

**L1-norm (Manhattan Distance)**  
$L_1 = \sum_i^n {|d_i|}$  

**L3-norm**  
$L_3 = \sqrt[3] {\sum_{i=1}^{n} {|d_i|}^3}$
