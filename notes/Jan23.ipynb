# Web Scraping with Pandas

```python
import pandas as pd
import requests
url = "https://en.wikipedia.org/wiki/The_Lord_of_the_Rings_(film_series)"
headers = {"User-Agent" : "Mozilla/5.0 (WIndows NT 10.0; Win64; x64)"}
response = requests.get(url, headers=headers)
print(response.text)
# lotr = pd.read_html(url)
```

```python
data = response.text
tables = pd.read_html(data) # Gives back an array of tables
display(tables[5])
```

```python
lotr_soundtrack = tables[3]
lotr_box_office = tables[4]
lotr_ratings = tables[5]

new_col_names = []

for col in lotr_box_office.columns:
    new_col_names.append(col[2]) # Get the third element of the tuple

lotr_box_office.columns = new_col_names
```

```python

display(lotr_box_office)
display(lotr_soundtrack)
display(lotr_ratings)
```

## Read tables from HTML using BeautifulSoup

```python
!pip install bs4
```

```python
import requests
from bs4 import BeautifulSoup
url = "https://en.wikipedia.org/wiki/The_Lord_of_the_Rings_(film_series)"

headers = {"User-Agent" : "Mozilla/5.0 (WIndows NT 10.0; Win64; x64)"}
response = requests.get(url, headers=headers)

data = response.text

soup = BeautifulSoup(data, 'lxml')
prettified_data = soup.prettify()
print(prettified_data)
```

```python
prettified_data.find_all('b')
```

```python
soup.find_all('table')[5]
```
